## 2. 합계/평균 계산기

## 4.단위 변환기


문제:
cm를 입력받아 inch로 변환 (1 inch = 2.54 cm)


```python
def cm_to_inch(cm_value):
    """
    Converts a centimeter value to inches.
    1 inch = 2.54 cm
    """
    # 1 cm = 1 / 2.54 inch
    inch_value = cm_value / 2.54
    return inch_value

# Get CM value from user input
input_cm = float(input("변환할 CM 값을 입력하세요: "))

# Perform conversion
converted_inch = cm_to_inch(input_cm)

# Print the result, formatted to two decimal places
print(f"{input_cm} cm는 {converted_inch:.2f} inch 입니다.")
```

## 단위 변환기 (CM-Inch, Celsius-Fahrenheit 선택)

```python
# 섭씨(Celsius)를 화씨(Fahrenheit)로 변환하는 함수를 정의합니다.
def celsius_to_fahrenheit(celsius_value):
    """
    Converts a Celsius value to Fahrenheit.
    Fahrenheit = (Celsius * 9/5) + 32
    """
    fahrenheit_value = (celsius_value * 9/5) + 32
    return fahrenheit_value

# 사용자에게 어떤 변환을 수행할지 선택하도록 안내합니다.
print("어떤 변환을 수행하시겠습니까?")
print("1. CM -> Inch")
print("2. Celsius -> Fahrenheit")
choice = input("선택(1 또는 2)을 입력하세요: ")

if choice == '1':
    # CM를 Inch로 변환하는 경우
    input_cm = float(input("변환할 CM 값을 입력하세요: "))
    converted_inch = cm_to_inch(input_cm)
    print(f"{input_cm} cm는 {converted_inch:.2f} inch 입니다.")
elif choice == '2':
    # Celsius를 Fahrenheit로 변환하는 경우
    input_celsius = float(input("변환할 섭씨(Celsius) 온도를 입력하세요: "))
    converted_fahrenheit = celsius_to_fahrenheit(input_celsius)
    print(f"{input_celsius}°C는 {converted_fahrenheit:.2f}°F 입니다.")
else:
    print("잘못된 선택입니다. 1 또는 2를 입력해주세요.")
```

### 테스트 케이스 1: 0°C 변환 (정확히 32°F)

```python
print("테스트 케이스 1: 0°C")
input_celsius_test1 = 0
converted_fahrenheit_test1 = celsius_to_fahrenheit(input_celsius_test1)
print(f"{input_celsius_test1}°C는 {converted_fahrenheit_test1:.2f}°F 입니다.")
```

```python
print("테스트 케이스 2: 100°C")
input_celsius_test2 = 100
converted_fahrenheit_test2 = celsius_to_fahrenheit(input_celsius_test2)
print(f"{input_celsius_test2}°C는 {converted_fahrenheit_test2:.2f}°F 입니다.")
```

## 8. 성적 분석 프로그램 만들기


### 문제 : 점수 5 개를 입력받아 리스트로 저장
출력: 평균/최대/최소/합계

도전: 90점 이상 개수 카운트하여 출력

"테스트 입력”을 함께 만들기


```python
# TODO: 성적 분석 프로그램 만들기 (함수 사용 안 함, for 문 사용)

# 사용자로부터 점수를 입력받아 리스트에 저장
# 'q'를 입력하면 입력을 종료합니다.
input_scores = []
while True:
    score_input = input("점수를 입력하세요 (종료하려면 'q' 입력): ")
    if score_input.lower() == 'q':
        break
    try:
        score = int(score_input)
        if 0 <= score <= 100:
            input_scores.append(score)
        else:
            print("점수는 0에서 100 사이여야 합니다. 다시 입력해주세요.")
    except ValueError:
        print("잘못된 입력입니다. 숫자를 입력하거나 'q'를 입력하세요.")

# 성적 분석 로직
if not input_scores:
    print("분석할 점수가 없습니다.")
else:
    # 합계 계산 (for 문 사용)
    total_sum = 0
    for score in input_scores:
        total_sum += score

    # 평균 계산
    average = total_sum / len(input_scores)

    # 최대 점수 계산 (for 문 사용)
    maximum = input_scores[0]
    for score in input_scores[1:]:
        if score > maximum:
            maximum = score

    # 최소 점수 계산 (for 문 사용)
    minimum = input_scores[0]
    for score in input_scores[1:]:
        if score < minimum:
            minimum = score

    # 90점 이상인 점수 개수 계산
    count_above_90 = sum(1 for score in input_scores if score >= 90)

    print(f"\n--- 성적 분석 결과 ---")
    print(f"합계: {total_sum}")
    print(f"평균: {average:.2f}")
    print(f"최대 점수: {maximum}")
    print(f"최소 점수: {minimum}")
    print(f"90점 이상인 점수 개수: {count_above_90}개")
```

add: 입력받아 append

list: 현재 목록 출력(번호 포함)

done: 번호를 입력받아 pop/remove로 삭제


```python
# TODO: TODO 리스트 만들기

todo_list = []

while True:
    print("\n--- ToDo 리스트 관리 ---")
    print("1. 할 일 추가")
    print("2. 할 일 목록 보기")
    print("3. 할 일 삭제")
    print("4. 종료")

    choice = input("선택하세요 (1-4): ")

    if choice == '1':
        task = input("추가할 할 일을 입력하세요: ")
        todo_list.append(task)
        print(f"'{task}'가 추가되었습니다.")
    elif choice == '2':
        if not todo_list:
            print("할 일이 없습니다.")
        else:
            print("\n--- 할 일 목록 ---")
            for i, task in enumerate(todo_list):
                print(f"{i + 1}. {task}")
    elif choice == '3':
        if not todo_list:
            print("삭제할 할 일이 없습니다.")
            continue
        print("\n--- 할 일 목록 ---")
        for i, task in enumerate(todo_list):
            print(f"{i + 1}. {task}")
        try:
            task_num = int(input("삭제할 할 일 번호를 입력하세요: "))
            if 1 <= task_num <= len(todo_list):
                removed_task = todo_list.pop(task_num - 1)
                print(f"'{removed_task}'가 삭제되었습니다.")
            else:
                print("잘못된 번호입니다. 다시 시도해주세요.")
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")
    elif choice == '4':
        print("ToDo 리스트 프로그램을 종료합니다.")
        break
    else:
        print("잘못된 선택입니다. 1에서 4 사이의 숫자를 입력해주세요.")
```

공식: Displacement = L * B * T * Cb * 1.025 (단, 방형비 Cb 는 0.8로 가정)

```python
# TODO: 선박 제원 출력 및 배수량 계산

# 사용자로부터 선박 정보 입력받기
ship_name = input("선박 이름을 입력하세요: ")
length = float(input("길이 (L)를 입력하세요 (미터): "))
breadth = float(input("폭 (B)을 입력하세요 (미터): "))
draft = float(input("흘수 (T)를 입력하세요 (미터): "))

# 방형비 Cb (상수)
Cb = 0.8

# 배수량 계산 공식: Displacement = L * B * T * Cb * 1.025
displacement = length * breadth * draft * Cb * 1.025

# 결과 출력
print(f"\n--- {ship_name} 선박 제원 및 배수량 ---")
print(f"선박 이름: {ship_name}")
print(f"길이 (L): {length:.2f} m")
print(f"폭 (B): {breadth:.2f} m")
print(f"흘수 (T): {draft:.2f} m")
print(f"계산된 배수량 (Displacement): {displacement:.2f} 톤")
```

### 03.  강재 리스트 반복 출력 및 합계 [반복문]

문제: 리스트에 저장된 여러 개의 강재 무게 데이터를 반복문을 통해 하나씩 출력하고, 전체 강재의 총 무게를 계산하는 프로그램을 작성하세요.

```python
# TODO: 강재 리스트 반복 출력 및 합계

# 강재 무게 데이터 리스트
steel_weights = [15.5, 23.0, 48.2, 7.8, 30.1, 55.0, 12.3]

print("--- 개별 강재 무게 ---")
total_weight = 0

# 리스트의 각 강재 무게를 반복하여 출력하고 총 무게에 더합니다.
for weight in steel_weights:
    print(f"{weight:.1f} 톤")
    total_weight += weight

print("\n--- 전체 강재 총 무게 ---")
print(f"총 강재 무게: {total_weight:.1f} 톤")
```
